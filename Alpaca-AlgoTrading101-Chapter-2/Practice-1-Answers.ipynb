## Question 1a Answers

Enter a buy market order for Roku (ROKU) if Netflix (NFLX) is up more than 5% in the last 5 days.

```python
ticker = 'ROKU'
reference_ticker = 'NFLX'

order_sent = None

while not order_sent:

    ref_barset = api.get_barset(reference_ticker, 'day', limit = 6).df[reference_ticker]
    ref_today_close = ref_barset['close'][-1]
    ref_five_days_ago_close = ref_barset['close'][-6]
    ref_percent_change = (ref_today_close - ref_five_days_ago_close)/ref_five_days_ago_close*100

    order_info = None

    print('Checking Trading Rule...')
    
    if ref_percent_change > 5:
        order_info = api.submit_order(
        symbol=ticker,
        side='buy',
        type='market',
        qty='1',
        time_in_force='day',
        )
        
        if  order_info.status == 'accepted':
            order_sent = True
    
    print(ref_percent_change) # Optional
    time.sleep(10)
    
print(order_info.status)
```

```python
ticker = 'ROKU'
reference_ticker = 'NFLX'

order_sent = None

while not order_sent:

    ref_barset = api.get_barset(reference_ticker, 'day', limit = 6).df[reference_ticker]
    ref_today_close = ref_barset['close'][-1]
    ref_five_days_ago_close = ref_barset['close'][-6]
    ref_percent_change = (ref_today_close - ref_five_days_ago_close)/ref_five_days_ago_close*100

    ticker_barset = api.get_barset(ticker, 'day', limit = 1).df[ticker]
    ticker_today_close = ticker_barset['close'][0]
    
    order_info = None

    print('Checking Trading Rule...')
    
    if ref_percent_change > 5:
        order_info = api.submit_order(
        symbol=ticker,
        side='buy',
        type='limit',
        limit_price=round(ticker_today_close * 0.99, 2), # Round to 2 decimal places
        qty='1',
        time_in_force='day',
        )
        
        if  order_info.status == 'accepted':
            order_sent = True
    
    print(ref_percent_change) # Optional
    time.sleep(3)
    
print(order_info.status)
```

## Question 2a Answers

Enter a buy limit order for Box (BOX) if Dropbox (DBX) is up more than 5% than BOX in the last 5 days AND BOX's volume yesterday is above the average volume of the 4 days before yesterday. Set the order limit price at 1% lower than the current BOX price.

```python

ticker = 'BOX'
reference_ticker = 'DBX'

order_sent = None

while not order_sent:

    ref_barset = api.get_barset(reference_ticker, 'day', limit = 6).df[reference_ticker]
    ref_today_close = ref_barset['close'][-1]
    ref_five_days_ago_close = ref_barset['close'][-6]
    ref_percent_change = (ref_today_close - ref_five_days_ago_close)/ref_five_days_ago_close*100

    ticker_barset = api.get_barset(ticker, 'day', limit = 6).df[ticker]
    ticker_today_close = ticker_barset['close'][-1]
    ticker_five_days_ago_close = ticker_barset['close'][-6]
    ticker_percent_change = (ticker_today_close - ticker_five_days_ago_close)/ticker_five_days_ago_close*100
    ticker_5d_average_volume = ticker_barset['volume'][:4].mean()
    ticker_ytd_volume = ticker_barset['volume'][-2]    
    
    diff_percent_change = ref_percent_change - ticker_percent_change
    
    order_info = None

    print('Checking Trading Rule...')
    
    if diff_percent_change > 5 and ticker_ytd_volume > ticker_5d_average_volume:
        order_info = api.submit_order(
        symbol=ticker,
        side='buy',
        type='limit',
        limit_price=round(ticker_today_close * 0.99, 2), # Round to 2 decimal places
        qty='1',
        time_in_force='day',
        )
        
        if  order_info.status == 'accepted':
            order_sent = True
    
    print(diff_percent_change) # Optional
    print(ticker_ytd_volume) # Optional
    print(ticker_5d_average_volume) # Optional
    time.sleep(3)
    
print(order_info.status)
```

# Question 3a:

We look at Carnival Corporation (CCL), Royal Caribbean Cruises (RCL) and Norwegian Cruise Line (NCLH).

If over the last 5 days, either one of the 3 stocks moves (i.e. rise or fall) 5% less than the other 2, we send in buy limit order for it. Set the order limit price at 10 cents higher than the current ask price.

E.g.

- Over the last 5 days, CCL and NCLH rose by 10% each, and RCL rose by 4%. We enter a buy limit order for RCL.
- Over the last 5 days, RCL and CCL fell by 2% each, and NCLH fell by 7%. We enter a buy limit order for NCLH.
- Over the last 5 days, NCLH rose by 6%, CCL rose by 2%, and RCL fell by 1%. We do nothing.

```python

import pandas as pd

ticker1 = 'CCL'
ticker2 = 'RCL'
ticker3 = 'NCLH'

order_sent = None

while not order_sent:

    ticker1_barset = api.get_barset(ticker1, 'day', limit = 6).df[ticker1]
    ticker1_today_close = ticker1_barset['close'][-1]
    ticker1_five_days_ago_close = ticker1_barset['close'][-6]
    ticker1_percent_change = (ticker1_today_close - ticker1_five_days_ago_close)/ticker1_five_days_ago_close*100
    ticker1_ask = api.get_last_quote(ticker1).askprice
    
    ticker2_barset = api.get_barset(ticker2, 'day', limit = 6).df[ticker2]
    ticker2_today_close = ticker2_barset['close'][-1]
    ticker2_five_days_ago_close = ticker2_barset['close'][-6]
    ticker2_percent_change = (ticker2_today_close - ticker2_five_days_ago_close)/ticker2_five_days_ago_close*100
    ticker2_ask = api.get_last_quote(ticker2).askprice
    
    ticker3_barset = api.get_barset(ticker3, 'day', limit = 6).df[ticker3]
    ticker3_today_close = ticker3_barset['close'][-1]
    ticker3_five_days_ago_close = ticker3_barset['close'][-6]
    ticker3_percent_change = (ticker3_today_close - ticker3_five_days_ago_close)/ticker3_five_days_ago_close*100
    ticker3_ask = api.get_last_quote(ticker3).askprice
        
    # Create lists for data we need. We will use these to create a dataframe    
    ticker_list = [ticker1, ticker2, ticker3]
    ticker_close_prices = [ticker1_today_close, ticker2_today_close, ticker3_today_close]
    ticker_ask_prices = [ticker1_ask, ticker2_ask, ticker3_ask]
    five_day_returns_list = [ticker1_percent_change, ticker2_percent_change, ticker3_percent_change]

    # Create dataframe with the data we need
    ticker_returns = pd.DataFrame(
                            {'Ticker': ticker_list,
                             'Close Prices': ticker_close_prices,
                             'Ask Prices': ticker_ask_prices,
                             '5 Day Returns (%)': five_day_returns_list
                            })
    
    ticker_returns.sort_values(by=['5 Day Returns (%)'], ascending=False, inplace=True) # Sort by returns. HIGHEST at the top.
    
    order_info = None

    print('Checking Trading Rule...')
    
    if ticker_returns.iloc[0,3] - ticker_returns.iloc[1,3] > 5 and ticker_returns.iloc[0,3] - ticker_returns.iloc[2,3] > 5:
        order_info = api.submit_order(
            symbol=ticker_returns.iloc[2,0],
            side='buy',
            type='limit',
            limit_price=ticker_returns.iloc[2,2] + 0.10, # Ask price
            # limit_price=ticker_returns.iloc[2,1] + 0.10, # Use current close price if ask prices are incorrect .
            qty='1',
            time_in_force='day',
        )
        
        if  order_info.status == 'accepted':
            order_sent = True
    
    print(ticker_returns) # Optional
    time.sleep(3)
    
print(order_info.status)
```

# Question 3b:

We look at Carnival Corporation (CCL), Royal Caribbean Cruises (RCL) and Norwegian Cruise Line (NCLH) again.

If over the last 5 days, either one of the 3 stocks moves 5% more than the other 2, we enter sell limit order for it. Set the order limit price at 10 cents lower than the current bid price.

Here, we are shorting the stock that didn't fall as much as the rest.

---

The general idea for question 3 is that we find a group of stocks that historically move together. If one of them moves different, we fire a trade in hopes that it this outlier will move similarly to the rest.

Note that this convergence might not happen if there is a change in the fundamentals of the outlier stock.

Some stocks as as biomedical stocks are less likely to converge if there is a big news event. Bio stocks are known to be very volatile. An approved patent or regulatory license can skyrocket a stock instantly, this this spike might not retract (so soon). Thus, it might not be a good idea to use bio stocks or stocks industries that tend to trend after a giant spike.

```python

import pandas as pd

ticker1 = 'CCL'
ticker2 = 'RCL'
ticker3 = 'NCLH'

order_sent = None

while not order_sent:

    ticker1_barset = api.get_barset(ticker1, 'day', limit = 6).df[ticker1]
    ticker1_today_close = ticker1_barset['close'][-1]
    ticker1_five_days_ago_close = ticker1_barset['close'][-6]
    ticker1_percent_change = (ticker1_today_close - ticker1_five_days_ago_close)/ticker1_five_days_ago_close*100
    ticker1_bid = api.get_last_quote(ticker1).bidprice
    
    ticker2_barset = api.get_barset(ticker2, 'day', limit = 6).df[ticker2]
    ticker2_today_close = ticker2_barset['close'][-1]
    ticker2_five_days_ago_close = ticker2_barset['close'][-6]
    ticker2_percent_change = (ticker2_today_close - ticker2_five_days_ago_close)/ticker2_five_days_ago_close*100
    ticker2_bid = api.get_last_quote(ticker2).bidprice
    
    ticker3_barset = api.get_barset(ticker3, 'day', limit = 6).df[ticker3]
    ticker3_today_close = ticker3_barset['close'][-1]
    ticker3_five_days_ago_close = ticker3_barset['close'][-6]
    ticker3_percent_change = (ticker3_today_close - ticker3_five_days_ago_close)/ticker3_five_days_ago_close*100
    ticker3_bid = api.get_last_quote(ticker3).bidprice
        
    # Create lists for data we need. We will use these to create a dataframe 
    ticker_list = [ticker1, ticker2, ticker3]
    ticker_close_prices = [ticker1_today_close, ticker2_today_close, ticker3_today_close]
    ticker_bid_prices = [ticker1_bid, ticker2_bid, ticker3_bid]
    five_day_returns_list = [ticker1_percent_change, ticker2_percent_change, ticker3_percent_change]

    # Create dataframe with the data we need
    ticker_returns = pd.DataFrame(
                            {'Ticker': ticker_list,
                             'Close Prices': ticker_close_prices,
                             'Bid Prices': ticker_bid_prices,
                             '5 Day Returns (%)': five_day_returns_list
                            })
    
    ticker_returns.sort_values(by=['5 Day Returns (%)'], ascending=True, inplace=True) # Sort by returns. LOWEST at the top.
    
    order_info = None

    print('Checking Trading Rule...')
    
    if ticker_returns.iloc[0,3] - ticker_returns.iloc[1,3] < 0 and ticker_returns.iloc[0,3] - ticker_returns.iloc[2,3] < 0 :
        order_info = api.submit_order(
            symbol=ticker_returns.iloc[2,0],
            side='sell',
            type='limit',
            limit_price=ticker_returns.iloc[2,2] - 0.10, # Bid price
            # limit_price=ticker_returns.iloc[2,1] + 0.10, # Use current close price if bid prices are incorrect.
            qty='1',
            time_in_force='day',
        )
        
        if  order_info.status == 'accepted':
            order_sent = True
    
    print(ticker_returns) # Optional
    time.sleep(3)
    
print(order_info.status)
```
