```python
# Requires installation
import quandl 
# library for extracting data from quandl.com 
# See: https://anaconda.org/anaconda/quandl

quandl.ApiConfig.api_key = 'insert api key' ### Insert your own API key. Create an account at quandl.com 

# More info on retrieving data from Quandl: https://www.quandl.com/tools/python

stockA = quandl.get_table('WIKI/PRICES', ticker = ['AAPL'], 
                        qopts = { 'columns': ['date', 'adj_close'] }, 
                        date = { 'gte': '2016-01-02', 'lte': '2020-12-31' }, 
                        paginate=True)

stockA
```
